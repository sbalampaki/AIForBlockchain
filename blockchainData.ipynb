## Getting the raw files for smart contracts

```python
import os
import glob
import pandas as pd
from huggingface_hub import snapshot_download

# downloading the dataset repo locally
local_dir = snapshot_download(
    repo_id="mwritescode/slither-audited-smart-contracts",
    repo_type="dataset",
    local_dir="slither_audited_contracts",
    local_dir_use_symlinks=False,
)


parquet_files = glob.glob(os.path.join(local_dir, "**", "*.parquet"), recursive=True)
print(f"Found {len(parquet_files)} parquet files")
print(parquet_files[:5])

#loading into dataframes
dfs = [pd.read_parquet(p) for p in parquet_files]
df = pd.concat(dfs, ignore_index=True)
```

```python
# can also be done on cli - evm-cfg-builder bytecode.txt
# this will generate .dot files
import subprocess
from pathlib import Path

def evm_cfg_builder(bytecode_path: str):
    bytecode_path = Path(bytecode_path).resolve()

    cmd = ["evm-cfg-builder", str(bytecode_path)]
    print("Running:", " ".join(cmd))

    result = subprocess.run(
        cmd,
        capture_output=True,
        text=True
    )

    print(result.stdout)
    if result.stderr:
        print(result.stderr)

    if result.returncode != 0:
        raise RuntimeError("evm-cfg-builder failed")

if __name__ == "__main__":
    evm_cfg_builder("bytecode.txt")
```

```python
# to generate cfgs
# command for terminal: find . -name "*.dot" -maxdepth 2 -print for f in $(find . -name "*.dot"); do dot -Tpng "$f" -o "${f%.dot}.png"; done
from pathlib import Path
import subprocess

root = Path(".").resolve()

dot_files = []
for path in root.rglob("*.dot"):
    depth = len(path.relative_to(root).parts)
    if depth <= 2:
        dot_files.append(path)
        print(path)

for dot_file in dot_files:
    png_file = dot_file.with_suffix(".png")
    subprocess.run(
        ["dot", "-Tpng", str(dot_file), "-o", str(png_file)],
        check=True
    )
    print(f"Generated {png_file}")

```
